```python
import pandas as pd 
```

```python
words = pd.read_csv("data/words_20000_clean.csv", index_col=0)
chars = words.iloc[:, 1:6]  # Get just the character columns
chars.head()
```

```python
# Get the frequency of each letter in that spot of the 5 spots 
freqs = pd.DataFrame()

char0 = chars['char0'].value_counts(normalize=True)
for char, freq in char0.iteritems():
    freqs.loc[char, 'char0'] = freq

char1 = chars['char1'].value_counts(normalize=True)
for char, freq in char1.iteritems():
    freqs.loc[char, 'char1'] = freq

char2 = chars['char2'].value_counts(normalize=True)
for char, freq in char2.iteritems():
    freqs.loc[char, 'char2'] = freq

char3 = chars['char3'].value_counts(normalize=True)
for char, freq in char3.iteritems():
    freqs.loc[char, 'char3'] = freq

char4 = chars['char4'].value_counts(normalize=True)
for char, freq in char0.iteritems():
    freqs.loc[char, 'char4'] = freq

freqs = freqs.sort_index()  # Sort the index a-z
freqs
```

```python
# Save the above data to a lookup table for use in Wordle solver
freqs.to_csv("data/lookup_char-freqs.csv")
```

```python
# Visualize the data (color coding doesn't show up on GitHub, fyi)
f = '{:.1f}' + ' %'
freqs_percs = freqs*100
freqs_styled = freqs_percs.style.apply(lambda x: ["background: red" if v == max(x) else "" for v in x], axis=0).format(f)
freqs_styled
```
